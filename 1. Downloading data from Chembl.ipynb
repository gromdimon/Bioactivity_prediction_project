```python
from chembl_webresource_client.new_client import new_client 
import pandas as pd
```

```python
target = new_client.target
target_query = target.search('Beta amyloid A4 prrotein')
targets = pd.DataFrame(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[3]
selected_target
```

```python
# Searching bioactivity data of this target
activity = new_client.activity
target_activity = activity.filter(target_chembl_id=selected_target).filter(standard_type='IC50')
act = pd.DataFrame.from_dict(target_activity)
act
```

```python
act.standard_type.unique()
```

```python
# Preparing data
act_normal = act[act.standard_value.notna()]
act_normal = act_normal[act_normal.canonical_smiles.notna()]
act_normal = act_normal.drop_duplicates(['canonical_smiles'])
act_normal
```

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
new_act = act_normal[selection]
bioactivity_threshold = []
for i in new_act.standard_value:
    if float(i) >= 10000:
        bioactivity_threshold.append('inactive')
    elif float(i) <= 1000:
        bioactivity_threshold.append('active')
    else:
        bioactivity_threshold.append('intermediate')
bioactivity_class = pd.Series(bioactivity_threshold, name = 'class')
new_act = pd.concat([new_act, bioactivity_class], axis=1)
new_act = new_act.dropna()
```
